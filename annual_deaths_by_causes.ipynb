```python
#importation du jeu de données
data = pd.read_csv("annual_deaths_by_causes.csv", delimiter = ",")
df = data.copy()
pd.set_option('display.max_column', 100)
pd.set_option('display.max_row', 100)
```

```python
df.info()
#Je remarque ici que la colonne "year" est considérée comme une colonne de nombres entiers (int64), je vais alors mettre à jour le format de la colonne au format texte.
```

```python
df["year"] = str(df["year"])
df.dtypes.value_counts()
df.dtypes.value_counts().plot.pie(title="Répartition des formats des colonnes", label='')
#on peut constater qu'il n'y a plus de colonne au format de nombres entiers. Et que la colonne "year" est maintenant considérée comme un objet.
```
